```python
import os
import pandas as pd
import numpy as np

path = 'C:/Users/Leand/OneDrive/Documentos/Lean/Analizador_imagenes_calcio/seleccion/010719SEL/'

dirlist = [ item for item in os.listdir(path) if os.path.isdir(os.path.join(path, item)) ]
for folder in dirlist:
    csv_list = os.listdir(path + folder + '/')
    csv_list = [x.split('.')[0] for x in csv_list if 'csv' in x if 'ratio' not in x]

    df = pd.DataFrame ()
    for table in csv_list:
        tabla = pd.read_csv(path + folder + '/' + table + '.csv',index_col=0, encoding='utf-8')
        tabla [table] = np.where(tabla['Pico_0'] < tabla['Pico_1'], tabla['Pico_1'], tabla['Pico_0'])
        df = pd.concat([df.reset_index(drop=True), tabla.reset_index(drop=True)], axis=1)
    df.to_csv(path + folder + '/' + 'ampratio_' + folder + '.csv',sep="\t", decimal=',')
```
