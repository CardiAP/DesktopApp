### Amplitud normalizada

```python
for i in range (0,len(file_tabulado)):
    try:
        cell = file_tabulado['célula'][i]
        photo = file_tabulado['foto'][i]
        wc = pd.read_csv(path + 'whole_cell/' + cell + '_analysis_result_' + photo + '.json.csv', sep = '\t', index_col=0)
        ampl = pd.read_csv(path + 'amplitudes/' + cell + '_analysis_result_' + photo + '.jsonamplitudes.csv', sep = '\t', index_col=0)
        local_min = pd.read_csv(path + 'min_peaks_intensities/' + cell + '_analysis_result_' + photo + '.jsonmin_peaks_intensities.csv', sep = '\t', index_col=0)
        df = pd.DataFrame()
        if file_tabulado['frecuencia'][i]==1:
            df['Transient1'] =  ampl['transient1']*local_min['transient1']/wc['Transient1'].iloc[3]
            df['Transient2'] =  ampl['transient2']*local_min['transient2']/wc['Transient2'].iloc[3]
            df.to_excel(path + 'amplitudes_norm/' + cell + '_' + photo + '.xlsx', engine='xlsxwriter')
        elif file_tabulado['frecuencia'][i]==5:
            file1 = file_tabulado[(file_tabulado['célula']==cell) & (file_tabulado['frecuencia']==1)]
            file1 = file1.reset_index(drop=True)
            wc1 = pd.read_csv(path + 'whole_cell/' + file1['célula'][0] + '_analysis_result_' + file1['foto'][0][:-3] + 'json.csv', sep = '[,\t]', index_col=0)
            df['Transient1'] =  ampl['transient1']*local_min['transient1']/wc1['Transient1'].iloc[3]
            df['Transient2'] =  ampl['transient2']*local_min['transient2']/wc1['Transient2'].iloc[3]
            df.to_excel(path + 'amplitudes_norm/' + cell + '_' + photo + '.xlsx', engine='xlsxwriter')
        else:
            pass
    except FileNotFoundError:
        pass
```
